```python
data_list = []
for ticker in tickers[1:]:
    url = URL_BASE + ticker + URL_ENDING
    session = requests.Session()
    response = session.get(url, headers=headers)
    response.encoding = "utf-8"
    soup = BeautifulSoup(response.text, "lxml")
    tables = soup.find_all("tbody", limit=2)
    result = []
    for table in tables:
        data = table.find_all("td")
        for i in data:
            result.append(i.text)
    ratios = []
    for i in range(1, len(result), 2):
        try:
            ratios.append(float(result[i].replace("\n", "").strip()))
        except Exception:
            result[i - 1] = 0
    dates = []
    for i in range(0, len(result), 2):
        if result[i] != 0:
            dates.append(str(result[i]))
    data_new = pd.DataFrame({"dates": dates, ticker: ratios})
    data_list.append(data_new)
    # try:
    #     data_base = pd.merge(data_base, data_new, how='left', on='dates')
    # except Exception:
    #     continue

# data_base = pd.concat(data_list, axis=1)
```
